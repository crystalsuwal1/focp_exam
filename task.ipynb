The Value of Learning Continuously

Why It Matters

# task 4


```python
def all_between_0_and_100(nums):
    for n in nums:
        if n < 0 or n > 100:
            return False
    return True

print(all_between_0_and_100([1, 52, 100]))   # True
print(all_between_0_and_100([9, 101]))      # False

```

```python
strng = []
while True:
    s = input("Enter a string (press Enter to stop): ")
    if s == "":
        break
    strng.append(s)

print("Unique strings entered:", len(set(strng)))
```
